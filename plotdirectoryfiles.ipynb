```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_datasets(directory_path):
    # List all .txt files in the given directory
    dataset_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if f.endswith('.txt')]

    # Iterate through each file and plot the data
    for dataset in dataset_files:
        # Load the data from the file
        data = np.loadtxt(dataset)

        # Extract columns: time, membrane potential, and current
        time = data[:, 0]
        membrane_potential = data[:, 1]
        current = data[:, 2]/1e11

        # Create a new figure for each dataset
        plt.figure()

        # Plot membrane potential vs time
        plt.plot(time, membrane_potential, label='Membrane Potential')

        # Plot current vs time (normalized for better visualization if necessary)
        plt.plot(time, current, label='Current', alpha=0.7)

        # Set the title with the filename
        plt.title(f"Data from {os.path.basename(dataset)}")

        # Add labels and legend
        plt.xlabel('Time')
        plt.ylabel('Values')
        plt.legend()

        # Display the plot
        plt.grid(True)
        plt.show()

# Example usage
directory_path = "datasetspulserisingpart"  # Replace with your directory path
plot_datasets(directory_path)
```
